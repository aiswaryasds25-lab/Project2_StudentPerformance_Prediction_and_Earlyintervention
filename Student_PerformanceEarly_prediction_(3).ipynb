```python
import pandas as pd

df = pd.read_csv('student-mat.csv', sep=';')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Create Pass/Fail column based on final grade (G3)
df['result'] = df['G3'].apply(lambda x: 'Pass' if x >= 10 else 'Fail')

df[['G3', 'result']].head()
```

```python
df['result'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['result'].value_counts().plot(kind='bar')
plt.title('Pass vs Fail Distribution')
plt.show()
```

```python
# Target
y = df['result']

# Features (drop final grade + result to avoid leakage)
X = df.drop(['G3', 'result'], axis=1)

X.head()
```

```python
import pandas as pd

X_encoded = pd.get_dummies(X, drop_first=True)

X_encoded.head()
```

```python
X_encoded.info()
```

```python
y = y.map({'Fail': 0, 'Pass': 1})

y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Create model
model = LogisticRegression(max_iter=1000)

# Train
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train, y_train)

y_pred_gb = gb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
```

```python
import pandas as pd

results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'Gradient Boosting'],
    'Accuracy': [
        accuracy_score(y_test, y_pred),
        accuracy_score(y_test, y_pred_dt),
        accuracy_score(y_test, y_pred_gb)
    ]
})

results
```

```python
results.set_index('Model').plot(kind='bar')
plt.title("Model Comparison")
plt.show()
```
